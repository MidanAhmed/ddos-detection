```python
import pandas as pd
import numpy as np
import random

data=pd.read_csv('dataset.csv')
result=list(data['pktcount']/data['dur'])
data['pktCount/dur']=result
result=list(data['bytecount']/data['dur'])
data['byteCount/dur']=result

# Replacing infinite with nan
data.replace([np.inf, -np.inf], np.nan, inplace=True)
  
# Dropping all the rows with nan values
data.dropna(inplace=True)

for i in data.index:
    if data.at[i, 'switch'] < random.randint(1, 10):
        data.at[i, 'switch'] += random.randint(1, 7)
    if data.at[i, 'dur'] < random.randint(1, 1881):
        data.at[i, 'dur'] += random.randint(1, 661)
    if data.at[i, 'pktcount'] < random.randint(53000, 200000):
        data.at[i, 'pktcount'] += random.randint(20000, 40000)
    if data.at[i, 'pktcount'] > random.randint(53000, 200000):
        data.at[i, 'pktcount'] -= random.randint(10000, 50000)
    if data.at[i, 'flows'] < random.randint(4, 10):
        data.at[i, 'flows'] += random.randint(1, 10)

s_dataset = data.groupby('label').get_group(0)
m_dataset = data.groupby('label').get_group(1)
s_dataset = s_dataset.drop(['dt','src','dst','label'], axis=1)
m_dataset = m_dataset.drop(['dt','src','dst','label'], axis=1)
s_dataset = pd.get_dummies(s_dataset)
m_dataset = pd.get_dummies(m_dataset)

def randomd(df, size):
    x = random.sample(list(df.index), k=size)
    d = df.loc[x]
    return d

def gen(m):
    m *= 200
    s = 20000-m
    m_data = randomd(m_dataset, m)
    s_data = randomd(s_dataset, s)
    fdata = pd.concat([m_data, s_data], ignore_index = True)
    fdata.to_csv(f"data{m}.csv", index=False)
    
gen(50)
gen(60)
gen(70)
data.describe()
```
