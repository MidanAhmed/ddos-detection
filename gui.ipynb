```python
%matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
from tkinter import *
import tkinter as tk
from tkinter import filedialog as fd
```

```python
win = Tk()
win.geometry("600x400")

labelframe1 = LabelFrame(win, text=" Step 1: Choose a CSV file to analyse. ")  
labelframe1.pack(fill="both", expand=True, padx=40)
    

def lr():
    def addlabels(x,y):
        for i in range(len(x)):
            plt.text(i, y[i], y[i], ha = 'center')
    model = joblib.load('logisticr.joblib')
    prediction = model.predict(data)
    fig, ax = plt.subplots(figsize =(5, 5))
    labels = ['Benign', 'Malicious']
    counts = [prediction.tolist().count(0), prediction.tolist().count(1)]
    bar_colors = ['tab:blue', 'tab:orange']
    ax.bar(labels, counts, color=bar_colors, label=labels, width=0.5)
    ax.set_ylabel('Number of packets')
    ax.set_title('Predicted Results for Logistic Regression')
    addlabels(labels, counts)
    plt.show()
    
def rf():
    def addlabels(x,y):
        for i in range(len(x)):
            plt.text(i, y[i], y[i], ha = 'center')
    model = joblib.load('randomf.joblib')
    prediction = model.predict(data)
    fig, ax = plt.subplots(figsize =(5, 5))
    labels = ['Benign', 'Malicious']
    counts = [prediction.tolist().count(0), prediction.tolist().count(1)]
    bar_colors = ['tab:blue', 'tab:orange']
    ax.bar(labels, counts, color=bar_colors, label=labels, width=0.5)
    ax.set_ylabel('Number of packets')
    ax.set_title('Predicted Results for Random Forest')
    addlabels(labels, counts)
    plt.show()
    
def dt():
    def addlabels(x,y):
        for i in range(len(x)):
            plt.text(i, y[i], y[i], ha = 'center')
    model = joblib.load('decisiont.joblib')
    prediction = model.predict(data)
    fig, ax = plt.subplots(figsize =(5, 5))
    labels = ['Benign', 'Malicious']
    counts = [prediction.tolist().count(0), prediction.tolist().count(1)]
    bar_colors = ['tab:blue', 'tab:orange']
    ax.bar(labels, counts, color=bar_colors, label=labels, width=0.5)
    ax.set_ylabel('Number of packets')
    ax.set_title('Predicted Results for Decision Tree')
    addlabels(labels, counts)
    plt.show()

def nb():
    def addlabels(x,y):
        for i in range(len(x)):
            plt.text(i, y[i], y[i], ha = 'center')
    model = joblib.load('naiveb.joblib')
    prediction = model.predict(data)
    fig, ax = plt.subplots(figsize =(5, 5))
    labels = ['Benign', 'Malicious']
    counts = [prediction.tolist().count(0), prediction.tolist().count(1)]
    bar_colors = ['tab:blue', 'tab:orange']
    ax.bar(labels, counts, color=bar_colors, label=labels, width=0.5)
    ax.set_ylabel('Number of packets')
    ax.set_title('Predicted Results for Gaussian Naive Bayes')
    addlabels(labels, counts)
    plt.show()
    
def ca():
    x = np.arange(4)
    y1 = []
    y2 = []
    model = joblib.load('logisticr.joblib')
    prediction = model.predict(data)
    y1.append(prediction.tolist().count(0))
    y2.append(prediction.tolist().count(1))
    model = joblib.load('randomf.joblib')
    prediction = model.predict(data)
    y1.append(prediction.tolist().count(0))
    y2.append(prediction.tolist().count(1))
    model = joblib.load('decisiont.joblib')
    prediction = model.predict(data)
    y1.append(prediction.tolist().count(0))
    y2.append(prediction.tolist().count(1))
    model = joblib.load('naiveb.joblib')
    prediction = model.predict(data)
    y1.append(prediction.tolist().count(0))
    y2.append(prediction.tolist().count(1))
    width = 0.40
    
    plt.bar(x-0.2, y1, width, color='tab:blue')
    plt.bar(x+0.2, y2, width, color='tab:orange')
    plt.xticks(x, ['Logistic Regression', 'Random Forest', 'Decision Tree', 'Gaussian NB'])
    plt.xlabel("Classifiers")
    plt.ylabel("Number of packets")
    plt.legend(["Benign", "Malicious"])
    plt.show()
    

def select_file():
    filetypes = (
        ("CSC files", "*.csv"),
        ('All files', '*.*')
    )

    filename = fd.askopenfilename(
        title='Open a file',
        initialdir='/Users/Midan/Minor Project',
        filetypes=filetypes)

    global data
    data = pd.read_csv(filename)
    slabel.config(text = f"Selected file: {filename}")


# open button
open_button = tk.Button(
    labelframe1,
    text='Open a File',
    command=select_file
)

open_button.pack(expand=True)

slabel = Label(labelframe1, text = "Selected file: ")
slabel.pack(expand=True)
  
labelframe2 = LabelFrame(win, text = " Step 2: Choose which algorithm to execute. ")  
labelframe2.pack(fill="both", expand = True, padx=40, pady=20)  
  
a = Button(labelframe2, text = "Logistic Regression", command=lr)
a.pack(expand=True)

b = Button(labelframe2, text = "Decision tree", command=dt)
b.pack(expand=True)

c = Button(labelframe2, text = "Random Forest", command=rf)
c.pack(expand=True)

d = Button(labelframe2, text = "Gaussian Naive Bayes", command=nb)
d.pack(expand=True)

e = Button(labelframe2, text = "Comparative Analysis", command=ca)
e.pack(expand=True)

win.mainloop()
```
